```python
#---- normalizing values to have common  

df2=df2.replace(['Rice'], 'rice')
df2=df2.replace(['Coconut'], 'coconut')
df2=df2.replace(['Jute'], 'jute')
df2=df2.replace(['Orange'], 'orange')
df2=df2.replace(['Cotton'], 'cotton')
df2=df2.replace(['Papaya'], 'papaya')
df2=df2.replace(['Apple'], 'apple')
df2=df2.replace(['Karbuja(Musk Melon)'], 'muskmelon')
df2=df2.replace(['Water Melon'], 'watermelon')
df2=df2.replace(['Grapes'], 'grapes')
df2=df2.replace(['Mango'], 'mango')
df2=df2.replace(['Banana'], 'banana')
df2=df2.replace(['Pomegranate'], 'pomegranate')
df2=df2.replace(['Lentil (Masur)(Whole)'], 'lentil')
df2=df2.replace(['Black Gram (Urd Beans)(Whole)'], 'blackgram')
df2=df2.replace(['Green Gram (Moong)(Whole)'], 'mungbean')
df2=df2.replace(['Pegeon Pea (Arhar Fali)'], 'pigeonpeas')
df2=df2.replace(['Bengal Gram Dal (Chana Dal)'], 'chickpea')
```

```python
final_df = pd.merge(df1, df2, on='label')
final_df
```

```python
#---- removing duplicates ----

print("Toal duplicate rows: ",final_df.duplicated().sum())
final_df.drop_duplicates()
```

```python
#----- checking for null values -----

final_df.columns[final_df.isna().any()].tolist()
```

```python
final_df.info()
final_df
```

```python
final_df = final_df.drop('arrival_date', axis=1)
final_df
```

```python
df_dummies = pd.get_dummies(final_df, prefix='crop', prefix_sep='.', columns=['label','state','district','market','variety'])
df_dummies
```

```python
#---- training the dataset ----

x = df_dummies.iloc[:,:-1].values
y = df_dummies.iloc[:,-1].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print("ACCURACY IN THE TRAINING MODEL =  ",training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy_rf = accuracy_score(x_test_prediction, y_test)
print("ACCURACY IN THE TEST MODEL =  ",test_data_accuracy_rf)
```
